```python
labels = ['Hostel', 'Hotel', 'Resort', 'Airbnb']
```

```python

# 사용자 정의 함수: 리스트 값을 그대로 반환
def func(pct, allvalues):
    absolute = int(pct/100.*sum(allvalues))
    return "{:.1f}%\n({:d}명)".format(pct, absolute)

# 파이 그래프 그리기
plt.figure(figsize=(8, 8))  # 그래프 크기를 더 크게 설정
plt.pie(house2, labels=labels, autopct=lambda pct: func(pct, house2), startangle=90, textprops={'fontsize': 14})

plt.title('7일 이하 여행자 숙소 유형', fontsize=16)
plt.show()
# 여행 기간 트렌드를 분석해보니 7일 여행이 가장 많았습니다.
# 그다음 8일 여행이 많았고
# 그다음 10일 이상 여행이 많았고 10일 여행 자수는 10명, 11일은 5명, 그 이후는 2명 밖에 없었기 때문에 하나로 합쳤습니다.
# 6일과 9일은 공동으로 많았습니다.
```

```python
# 여행 기간에 따른 숙박비와 교통비 지출량 분석
# 평균과 중앙값, 표준편차를 구해 여행 기간에 따른 지출 패턴을 파악
# 그룹화된 데이터를 이용해 각 여행 기간에 따른 지출의 분포를 시각화 하기
tso=tdf_seven_down[tdf_seven_down['Accommodation type'] == 'Hotel']
tdf_seven_over.info()
```

```python
import seaborn as sns
# 데이터 예시: tdf_five의 'Accommodation cost' 열
accommodation_cost = tso['Accommodation cost']

# 평균, 중앙값, 표준편차 계산
mean_value = accommodation_cost.mean()
median_value = accommodation_cost.median()
std_dev = accommodation_cost.std()

# 히스토그램 생성
plt.figure(figsize=(10, 6))
sns.histplot(accommodation_cost, bins=20, kde=True, color='skyblue')

# 평균, 중앙값, 표준편차를 표시
plt.axvline(mean_value, color='r', linestyle='--', label=f'평균값: {mean_value:.2f}')
plt.axvline(median_value, color='g', linestyle='-', label=f'중앙값: {median_value:.2f}')
plt.axvline(mean_value + std_dev, color='b', linestyle=':', label=f'평균 + 표준편차: {mean_value + std_dev:.2f}')
plt.axvline(mean_value - std_dev, color='b', linestyle=':', label=f'평균 - 표준편차: {mean_value - std_dev:.2f}')

# 레전드와 제목 설정
plt.legend()
plt.title('7일 이하 여행자 호텔이용 히스토그램')
plt.xlabel('숙박비')
plt.ylabel('빈도수')
```
